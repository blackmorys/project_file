```python
df['NA_Sales'] = df['NA_Sales'].str.replace(pat=r'[M]', repl=r'', regex = True)
df['EU_Sales'] = df['EU_Sales'].str.replace(pat=r'[M]', repl=r'', regex = True)
df['JP_Sales'] = df['JP_Sales'].str.replace(pat=r'[M]', repl=r'', regex = True)
df['Other_Sales'] = df['Other_Sales'].str.replace(pat=r'[M]', repl=r'', regex = True)
```

```python
def fix_columns(col):
    temp_list = []

    for i in col:
        temp = float(i.replace("K",""))
        temp_list.append(temp)

    return temp_list

les_na = df.loc[df.NA_Sales.str.contains('K')]
les_na
```

```python
les_na['NA_Sales'] = fix_columns(les_na['NA_Sales'])
les_na['NA_Sales'] = les_na['NA_Sales'] * 0.001
les_na = les_na['NA_Sales']
les_na
```

```python
les_eu = df.loc[df.EU_Sales.str.contains('K')]
les_eu
```

```python
les_eu['EU_Sales'] = fix_columns(les_eu['EU_Sales'])
les_eu['EU_Sales'] = les_eu['EU_Sales'] * 0.001
les_eu = les_eu['EU_Sales']
les_eu
```

```python
les_jp = df.loc[df.JP_Sales.str.contains('K')]
les_jp
```

```python
les_jp['JP_Sales'] = fix_columns(les_jp['JP_Sales'])
les_jp['JP_Sales'] = les_jp['JP_Sales'] * 0.001
les_jp = les_jp['JP_Sales']
les_jp
```

```python
les_ot = df.loc[df.Other_Sales.str.contains('K')]
les_ot
```

```python
les_ot['Other_Sales'] = fix_columns(les_ot['Other_Sales'])
les_ot['Other_Sales'] = les_ot['Other_Sales'] * 0.001
les_ot = les_ot['Other_Sales']
les_ot
```

```python
df['NA_Sales'].loc[df.NA_Sales.str.contains('K')] = les_na
df['EU_Sales'].loc[df.EU_Sales.str.contains('K')] = les_eu
df['JP_Sales'].loc[df.JP_Sales.str.contains('K')] = les_jp
df['Other_Sales'].loc[df.Other_Sales.str.contains('K')] = les_ot
df
```

```python
df = df.astype({'NA_Sales':'float', 'EU_Sales':'float', 'JP_Sales':'float', 'Other_Sales':'float'})
df.sum()
```
